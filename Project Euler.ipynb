```python
#Problem 23
#January 17, 2017

def divsum(x):
    divsum=1
    for i in range(2,int(x**.5)+1):
        if not x%i:
            if i==x/i:
                divsum+=i
            else:
                divsum+=i+x/i
    return divsum

def abnums(m):
    abund=[]
    for x in xrange(12,m):
        if divsum(x)>x:
            abund.append(x)
    return abund
start=timer()
n=28123
abn= abnums(n+1)
nums=np.ones(n+1,dtype=np.bool)
for a,b in enumerate(abn):
    for c,d in enumerate(abn[a:]):
        s=b+d
        if s<=n and nums[s]:
            nums[b+d]=False
print sum(np.nonzero(nums)[0]), timer()-start
```
